```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
%matplotlib Inline

image = cv2.imread("popitt.jpg")

image_copy=np.copy(image)
image_copy=cv2.cvtColor(image_copy,cv2.COLOR_BGR2RGB)
plt.imshow(image_copy)
```

```python
gray=cv2.cvtColor(image_copy,cv2.COLOR_RGB2GRAY)
grap=np.float32(gray)

#detect corners
dst=cv2.cornerHarris(gray,2,3,0.04)

dst=cv2.dilate(dst,None)

plt.imshow(dst,cmap='gray')
```

```python
thresh=0.01*dst.max()

corner_image=np.copy(image_copy)

for j in range(0,dst.shape[0]):
    for i in range(0,dst.shape[1]):
        if(dst[j,i]>thresh):
            cv2.circle(corner_image,(i,j),2,(0,255,0),2)
plt.imshow(corner_image)
```

```python
#sift
image1 = cv2.imread("popitt.jpg")
image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(gray, None)
image_with_keypoints = cv2.drawKeypoints(image1, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(8, 8))
plt.imshow(image_with_keypoints)
plt.axis('off')
plt.show()
```

```python
#surf
orb = cv2.ORB_create()
keypoints, descriptors = orb.detectAndCompute(gray, None)
image_with_keypoints = cv2.drawKeypoints(image1, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
plt.imshow(image_with_keypoints)

plt.axis('off')
plt.show()
```
